```python
from sklearn.datasets import load_iris
import pandas as pd
from pandas import DataFrame as df
```

<h1>PROBLEM 1</h1>

```python
iris_ds = load_iris()

x = df(iris_ds.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
y = df(iris_ds.target, columns=['Species'])
print(x)
print(y)
```

<h1>PROBLEM 2</h1>

```python
dataframe = pd.concat([x, y], axis=1)
dataframe
```

<h1>PROBLEM 3</h1>

```python
# Display the 4th sample from the beginning and see what each feature is like.
print("4th Sample:")
print(dataframe.head(4).iloc[3])  # Displaying the 4th sample (index 3)

# Output the total number of samples for each label.
print("\nTotal number of samples for each label:")
print(dataframe['Species'].value_counts())

# Check if there is a missing value in the feature quantity.
print("\nChecking for missing values:")
print(dataframe.isnull().sum())

# Display the mean, standard deviation, and quartiles of the feature values at once.
print("\nSummary Statistics:")
print(dataframe.describe())
```

<h1>PROBLEM 6</h1>

```python
import matplotlib.pyplot as plt
```

```python
#Make a pie chart of the number of samples per label (while showing percentages)
labels = ['0', '1', '2']
total = dataframe.Species.sum()
number0 = (dataframe['Species'].value_counts()[0]) * 100 / total
number1 = (dataframe['Species'].value_counts()[1]) * 100 / total
number2 = (dataframe['Species'].value_counts()[2]) * 100 / total
values = [number0, number1, number2]

plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=0)
plt.title('Pie Chart')
plt.show()
```

```python
#Select features one by one and visualize the distribution of data for each label using a box plot
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(15, 5))


dataframe.boxplot(column='sepal_length', ax=axes[0])
axes[0].set_title('Sepal Length')

dataframe.boxplot(column='sepal_width', ax=axes[1])
axes[1].set_title('Sepal Width')

dataframe.boxplot(column='petal_length', ax=axes[2])
axes[2].set_title('Petal Length')

dataframe.boxplot(column='petal_width', ax=axes[3])
axes[3].set_title('Petal Width')
```

```python
import seaborn as sns

#Select features one by one and visualize the distribution of data for each label using a violin plot
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(15, 5))


sns.violinplot(x='Species', y='sepal_length', data=dataframe, ax=axes[0])
axes[0].set_title('Sepal Length')

sns.violinplot(x='Species', y='sepal_width', data=dataframe, ax=axes[1])
axes[1].set_title('Sepal Width')

sns.violinplot(x='Species', y='petal_length', data=dataframe, ax=axes[2])
axes[2].set_title('Petal Length')

sns.violinplot(x='Species', y='petal_width', data=dataframe, ax=axes[3])
axes[3].set_title('Petal Width')


plt.suptitle('Violin Plots of Iris Dataset', y=1.02)
plt.tight_layout()
plt.show()
```
